```python
# Loading pandas library and it contains function related to the database
import pandas as pd
```

```python
# Reading the data in csv and assing it to a variable

dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
# Get dummies means to convert ordernal/nominal to numbers
# Here we convert nominal to numbers (one hot encoding method)

dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
# to get table title

dataset.columns
```

```python
# input / independent

independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]

# output / dependent
dependent = dataset[['Profit']]
```

```python
# Loading sklearn library to support train, test & algorithm
from sklearn.model_selection import train_test_split

# Spliting Training Set and Test Set
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Loading the DecisionTreeRegressor algorithm from sklearn
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion="absolute_error",splitter="best")
```

```python
# Learning Pharse is happening here
regressor.fit(X_train,Y_train)
```

```python
# To show as graph
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# Prediction for test dataset
y_pred = regressor.predict(X_test)
```

```python
# Loading the r2 from sklearn
from sklearn.metrics import r2_score
# checking r2 score here
r2_score = r2_score(Y_test,y_pred)
r2_score
```
