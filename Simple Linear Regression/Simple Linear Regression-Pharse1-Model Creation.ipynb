```python
import pandas as pd
```

```python
# Reading the data in csv and assing it to a variable
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#printing dataset
dataset
```

```python
# Set input from dataset & Printing it
independent =  dataset[["YearsExperience"]]
independent
```

```python
# Set output from dataset & Printing it
dependent =  dataset[["Salary"]]
dependent
```

```python
# Loading sklearn library to support train, test & algorithm
from sklearn.model_selection import train_test_split

# Spliting Training Set and Test Set
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Loading the linear regression algorithm from sklearn
from sklearn.linear_model import LinearRegression

# Assingning the fuction to variable
regressor = LinearRegression()
```

```python
# Learning Pharse is happening here
regressor.fit(X_train,Y_train)
```

```python
# Weight Checking
weight = regressor.coef_
weight
```

```python
# Bias Checking
bias = regressor.intercept_
bias
```

```python
# Prediction for test dataset
y_pred = regressor.predict(X_test)
```

```python
# Loading the r2 from sklearn
from sklearn.metrics import r2_score

# checking r2 score here
r2_score = r2_score(Y_test,y_pred)
```

```python
#printing r2 value here if it near to 1 model leaned good else more errors
r2_score
```

```python
# loading pickle to save final model
import pickle
```

```python
# declare file name here
filename = "finalized_model_linear.sav";
```

```python
# dump model to file and wb stands for "write binary"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# load model and test it and rb stands for "read binary"
loaded_model = pickle.load(open(filename,'rb'))
# check model by dummy data input
result = loaded_model.predict([[13]])
```
