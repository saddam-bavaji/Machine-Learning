```python
# Loading pandas library and it contains function related to the database
import pandas as pd
```

```python
# Reading the data in csv and assing it to a variable

dataset = pd.read_csv("insurance_pre.csv")
```

```python
# to remove one dummies from the table to make optimized we use like

dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
# to get table title

dataset.columns
```

```python
# input / independent

independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
# output / dependent
dependent = dataset[['charges']]
```

```python
# Loading sklearn library to support train, test & algorithm
from sklearn.model_selection import train_test_split
```

```python
# Spliting Training Set and Test Set
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Loading the linear regression algorithm from sklearn
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
# Learning Pharse is happening here
regressor.fit(X_train,Y_train)
```

```python
# Weight Checking
weight = regressor.coef_
weight
```

```python
# Bias Checking
bias = regressor.intercept_
bias
```

```python
# Prediction for test dataset
y_pred = regressor.predict(X_test)
```

```python
# Loading the r2 from sklearn
from sklearn.metrics import r2_score

# checking r2 score here
r2_score = r2_score(Y_test,y_pred)
r2_score
```
