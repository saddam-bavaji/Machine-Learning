```python
# Loading pandas library and it contains function related to the database
import pandas as pd
```

```python
# Reading the data in csv and assing it to a variable

dataset = pd.read_csv("insurance_pre.csv")
```

```python
# to remove one dummies from the table to make optimized we use like

dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
# input / independent

independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
# output / dependent
dependent = dataset[['charges']]
```

```python
# Loading sklearn library to support train, test & algorithm
from sklearn.model_selection import train_test_split
```

```python
# Spliting Training Set and Test Set
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Loading the  regression algorithm from sklearn
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=600,max_depth=10,min_samples_split=10,min_samples_leaf=5,max_features='log2')
```

```python
# Learning Pharse is happening here
regressor.fit(X_train,Y_train)
```

```python
# Prediction for test dataset
y_pred = regressor.predict(X_test)
```

```python
# Loading the r2 from sklearn
from sklearn.metrics import r2_score
# checking r2 score here
r2_score = r2_score(Y_test,y_pred)
r2_score
```

```python
# loading pickle to save final model
import pickle
```

```python
# declare file name here
filename = "assignment_finalized_model_random_forest.sav";
```

```python
# dump model to file and wb stands for "write binary"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# load model and test it and rb stands for "read binary"
loaded_model = pickle.load(open('assignment_finalized_model_random_forest.sav','rb'))

# we need to give input for testing base on how much input we have in training set
# so we give 5 inputs inside for testing

result = loaded_model.predict([[19,0,27.9,0,1]])
result
```
